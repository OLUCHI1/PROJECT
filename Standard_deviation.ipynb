```python
import numpy as np
import matplotlib as plt
import scipy.io as spio
import pandas as pd

%matplotlib inline 
```

```python
matdata = spio.loadmat('trajectories.mat')                         
matarray = matdata['trajectories04000415'] 

print(matarray[0])

data = matarray
```

```python
labels = ['Vehicle ID', 'Frame ID', 'Total Frames', 'Global Time', 'Local X', 'Local Y', 
          'Global X', 'Global Y', 'Length', 'Width', 'Class', 'Velocity', 'Acceleration', 
          'Lane ID', 'Preceding', 'Following', 'Spacing', 'Headway']
df = pd.DataFrame(data = matarray, columns = labels ) 
```

```python
df['Vehicle ID'] = df['Vehicle ID'].astype(int)
df['Frame ID'] = df['Frame ID'].astype(int)
df['Total Frames'] = df['Total Frames'].astype(int)
df['Class'] = df['Class'].astype(int)
df['Preceding'] = df['Preceding'].astype(int)
df['Following'] = df['Following'].astype(int)
df['Lane ID'] = df['Lane ID'].astype(int)

```

```python
# returns dataframe specific to vehicle ID

for i in range(1, 2911 + 1):
    def vehicle_df(id = i):
        return df.loc[df['Vehicle ID'] == id]

#calculate the standard deviation for the velocity of each vehicle
    def std_vel_by_vid(id = i):
        return vehicle_df(id)['Velocity'].std()

#identify the class of vehicle  
    def vehicle_class(id = i):
        return vehicle_df(id)['Class'].mean()
            
  
    print(i,std_vel_by_vid(),vehicle_class());

```
